```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
FILENAME = './data/laycock_edge_list.csv'
```

```python
Laydf = pd.read_csv(FILENAME, sep=',')
```

```python
Laydf
```

```python
img1 = Laydf.iloc[:6]
img2 = Laydf.iloc[6:12]
img3 = Laydf.iloc[12:18]
img4 = Laydf.iloc[18:24]
img5 = Laydf.iloc[24:30]

SCALE = 800
```

```python
G1 = nx.from_pandas_edgelist(img1,'node1','node2',edge_attr=None)
emotion = img1['node2'].tolist()
valuess = img1['val'].tolist()
img1_size = dict(zip(emotion, valuess))
img1_size['Image 1'] = 10
img1_sizels = [SCALE*img1_size[n] for n in G1.nodes()]
```

```python
pos = {'Image 1': ([0,0]),
 'Negative connotations': ([-0.5,-0.2]),
 'Positive connotations': ([0.8 , 0]),
 'Familiarity': ([0.5,-0.2]),
 'Personal connection': ([-0.5,0.2]),
 'Personal response shaped by location': ([0.5,0.2]),
 'Stated no emotion': ([-0.8, 0])}

plt.figure(figsize=(10,10))
plt.axis('off')
plt.title('Image 1',fontsize=18)
nx.draw_networkx_nodes(G1, pos, node_color='b', node_size=img1_sizels, alpha=.9)
nx.draw_networkx_edges(G1, pos, edge_color='k',width=1, alpha=.5)
nx.draw_networkx_labels(G1, pos, font_size=12, font_color='k')
```
